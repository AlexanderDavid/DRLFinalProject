```python
import pickle

rewards, lengths, losses = pickle.load(open("./results_checkpoint.pickle", "rb"))

rewards = rewards[30:]
lengths = lengths[30:]

rewards = list(sorted(rewards, key=lambda x: x[0]))
lengths = list(sorted(lengths, key=lambda x: x[0]))
```

```python
from matplotlib import pyplot as plt
import seaborn as sns

sns.set_theme(style="whitegrid")
sns.set_context("paper", font_scale=2)
plt.tight_layout()

lw = 3

sns.lineplot(x=[reward[0] for reward in rewards], y=[reward[1] for reward in rewards], linewidth=lw)
plt.savefig("report/figures/rewards.png")
plt.xlabel("Steps")
plt.ylabel("Reward")
plt.show()

sns.lineplot(x=range(len(lengths)), y=[length[1] for length in lengths], linewidth=lw)
plt.xlabel("Episodes")
plt.ylabel("Episode Length")
plt.savefig("report/figures/lengths.png")
plt.show()
```
